```python
from statistics import mode
```

```python
for column in df.columns:
    if column != 'SalePrice':
        if df[column].dtype == 'int' or df[column].dtype == 'float':
            column_mean = df[column].mean()
            df[column].fillna(column_mean, inplace=True)
        elif df[column].dtype == 'object':
            column_mode = mode(df[column].dropna())
            df[column].fillna(column_mode, inplace=True)
#eÄer kolonda sayÄ±lar varsa sayÄ±larÄ±n ortalamasÄ±nÄ± boÅluklara yerleÅtirdik, eÄer string deÄerler 
#varsa en Ã§ok tekrarlanan deÄeri boÅluklara yerleÅtirdik
```

```python
for column in df.columns:
    if column != 'SalePrice': #saleprice kolonu hariÃ§ her kolona data encoding yapÄ±yorum
        value_counts = df[column].value_counts()
        num_categories = len(value_counts)
        
        # Ä°ki alt kategorisi varsa 0 ve 1 olarak kodlayalÄ±m
        if num_categories == 2:
            category_mapping = {category: index for index, category in enumerate(value_counts.index)}
            df[column] = df[column].map(category_mapping)
        # Ä°ki alt kategoriden fazlaysa 1'den baÅlayan sÄ±ralÄ± sayÄ±larla kodlayalÄ±m
        else:
            category_mapping = {category: index + 1 for index, category in enumerate(value_counts.index)}
            df[column] = df[column].map(category_mapping)
```

```python
'''yaygÄ±n outlier tespit yÃ¶ntemi, IQR (Interquartile Range - Ãeyrekler ArasÄ± Menzil) yÃ¶ntemidir. 
IQR, veri kÃ¼mesinin alt Ã§eyrek ve Ã¼st Ã§eyrek arasÄ±ndaki mesafeyi ifade eder ve veri noktalarÄ±nÄ±n 
merkezsel yÃ¼zde 50'sini iÃ§erir. Bir veri noktasÄ±, alt Ã§eyrekten IQR'Ä±n 1.5 katÄ± daha kÃ¼Ã§Ã¼k veya Ã¼st
Ã§eyrekten IQR'Ä±n 1.5 katÄ± daha bÃ¼yÃ¼kse, bu veri noktasÄ± bir outlier olarak kabul edilir.'''

def remove_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_rows = ((df < lower_bound) | (df > upper_bound)).any(axis=1)
    df_filtered = df[~outlier_rows]
    return df_filtered

# Outlier'larÄ± tespit edelim ve veri kÃ¼mesinden Ã§Ä±karalÄ±m
df_without_outliers = remove_outliers(df)

```
