```python
from statistics import mode
```

```python
for column in df.columns:
    if column != 'SalePrice':
        if df[column].dtype == 'int' or df[column].dtype == 'float':
            column_mean = df[column].mean()
            df[column].fillna(column_mean, inplace=True)
        elif df[column].dtype == 'object':
            column_mode = mode(df[column].dropna())
            df[column].fillna(column_mode, inplace=True)
#eğer kolonda sayılar varsa sayıların ortalamasını boşluklara yerleştirdik, eğer string değerler 
#varsa en çok tekrarlanan değeri boşluklara yerleştirdik
```

```python
for column in df.columns:
    if column != 'SalePrice': #saleprice kolonu hariç her kolona data encoding yapıyorum
        value_counts = df[column].value_counts()
        num_categories = len(value_counts)
        
        # İki alt kategorisi varsa 0 ve 1 olarak kodlayalım
        if num_categories == 2:
            category_mapping = {category: index for index, category in enumerate(value_counts.index)}
            df[column] = df[column].map(category_mapping)
        # İki alt kategoriden fazlaysa 1'den başlayan sıralı sayılarla kodlayalım
        else:
            category_mapping = {category: index + 1 for index, category in enumerate(value_counts.index)}
            df[column] = df[column].map(category_mapping)
```

```python
'''yaygın outlier tespit yöntemi, IQR (Interquartile Range - Çeyrekler Arası Menzil) yöntemidir. 
IQR, veri kümesinin alt çeyrek ve üst çeyrek arasındaki mesafeyi ifade eder ve veri noktalarının 
merkezsel yüzde 50'sini içerir. Bir veri noktası, alt çeyrekten IQR'ın 1.5 katı daha küçük veya üst
çeyrekten IQR'ın 1.5 katı daha büyükse, bu veri noktası bir outlier olarak kabul edilir.'''

def remove_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_rows = ((df < lower_bound) | (df > upper_bound)).any(axis=1)
    df_filtered = df[~outlier_rows]
    return df_filtered

# Outlier'ları tespit edelim ve veri kümesinden çıkaralım
df_without_outliers = remove_outliers(df)

```
